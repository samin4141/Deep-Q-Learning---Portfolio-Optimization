```python
import yfinance as yf
import pandas as pd

# Define the tickers for the assets
tickers = {
    'Vanguard_Index_Fund': 'VFINX',
    'S&P_500_Index_Fund': '^GSPC',
    'Gold': 'GC=F',
    'Crude_Oil': 'CL=F',
    '10Y_Treasury_Bonds': '^TNX'
}

# Set the time period for the data (for example, last 5 years)
start_date = '1940-01-01'
end_date = '2023-01-01'

# Fetch data for each ticker
data = {}
for name, ticker in tickers.items():
    data[name] = yf.download(ticker, start=start_date, end=end_date)

# Show the data for Vanguard Index Fund as an example
print("Vanguard Index Fund data:")
print(data['Vanguard_Index_Fund'].head())

# Alternatively, you can save the data to CSV files for further analysis
#for name, df in data.items():
#    df.to_csv(f'{name}_data.csv')
#    print(f"Saved {name} data to {name}_data.csv")
```
